```python
spent_per_customer = orders.groupby(['CustomerID', 'Country']).agg({'amount_spent': 'sum'}).sort_values(by = 'amount_spent', ascending=False)
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
vip = spent_per_customer[spent_per_customer['amount_spent']>spent_per_customer['amount_spent'].quantile(0.95)]
vip
```

```python
preferred = spent_per_customer[(spent_per_customer['amount_spent']>spent_per_customer['amount_spent'].quantile(0.75))& (spent_per_customer['amount_spent']<spent_per_customer['amount_spent'].quantile(0.95))]
preferred
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
vip
```

```python
vip = vip.drop(columns = 'index')
```

```python
preferred = preferred.reset_index()
```

```python
vip['Country'].value_counts()
```

```python
combined = pd.concat([vip, preferred])
combined['Country'].value_counts()[0:1]
```
